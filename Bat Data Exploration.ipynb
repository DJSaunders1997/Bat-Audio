```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys
```

```python
# Plot first 3 files
# Amplitude the same for all plots


for file in range(12890950, 12890952+1):
    
    print()
    
    spf = wave.open("Wav-and-XML-subsample/{}.wav".format(file), "r")

    # Extract Raw Audio from Wav File
    signal = spf.readframes(-1)
    signal = np.fromstring(signal, "Int16")
    fs = spf.getframerate()
    
    Time = np.linspace(0, len(signal) / fs, num=len(signal))

    plt.figure(figsize=(25,10))
    plt.ylim((-30000,30000))
    #plt.xlim((0,3))
    plt.ylabel("Amplitude (?)")
    plt.xlabel("Time (s)")

    plt.figure(1)
    plt.title('Signal wave of {}.wav'.format(file))
    plt.plot(Time, signal)
    plt.show()

```

```python
#File 1

spf = wave.open("Wav-and-XML-subsample/12890950.wav", "r")

# Extract Raw Audio from Wav File
signal = spf.readframes(-1)
signal = np.fromstring(signal, "Int16")
fs = spf.getframerate()



print(spf.getnchannels())
Time = np.linspace(0, len(signal) / fs, num=len(signal))

plt.figure(figsize=(25,10))
plt.ylim((-30000,30000))
plt.xlim((0,3))

plt.figure(1)
plt.title("Signal Wave...")
plt.plot(Time, signal)
plt.show()
```
