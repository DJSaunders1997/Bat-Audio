```python
# Imports
from scipy.io.wavfile import read
import matplotlib.pyplot as plt
```

```python
#https://stackoverflow.com/a/25095962
# read audio samples
input_data = read("Wav-and-XML-subsample/12890950.wav")
audio = input_data[1]
# plot the first 1024 samples
plt.plot(audio[0:1024])
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time")
# set the title  
plt.title("Sample Wav")
# display the plot
plt.show()
```

```python
# read audio samples
input_data = read("Wav-and-XML-subsample/12890950.wav")
audio = input_data[1]
print(audio.size)

# plot the first 1024 samples
plt.plot(audio[0:audio.size])
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time")
# set the title  
plt.title("Sample Wav")
# display the plot
plt.show()
```
